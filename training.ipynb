```python
from ultralytics import YOLO

model = YOLO('yolov8n.pt')


results = model.train(
    data='medicine_final/data.yaml',
    epochs=200,
    imgsz=640,
    degrees=15,                  
    translate=0.1,               
    scale=0.1,                   
    fliplr=0.5,                  
    hsv_s=0.7,                  
    hsv_v=0.4                    
)
```
